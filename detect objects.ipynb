```python
# pip install opencv-python ultralytics
import cv2  # المكتبه اللي بنستخدمها عشان نتعامل مع الصور 
from ultralytics import YOLO   # موديل اليولو اللي بيعمل ديتيكت للوبجيكتس

model = YOLO(r"C:\Users\Admin\Downloads\best.pt")  # بجيب الموديل بتاع اليولو اللي هشتغل بيه من الفايل اللي نزلناها لان ده موديل مدرب و جاهز يعني بستدعي الموديل الجاهز

cap = cv2.VideoCapture(r"C:\Users\Admin\Downloads\TrafficPolice.mp4")  # اقراء الفيديو عن طريق سي في 2 لان ده اللي بيتعامل مع الفيديوهات و الصور و بيعرف يقرأها
fps = int(cap.get(cv2.CAP_PROP_FPS))        # هات عدد الفريمات اللي هتشتغل عليها
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))      # هات عرض الفريمات اللي هتشتغل عليها
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))        # هاتارتفاع الفريمات اللي هتشتغل عليها ليه بعمل كده علشان هو هيعمل فيديو جديد فلازم يكون معاه بيانات الفيديو القديم لانه هيظبط البوكسيز علي نفس البيانات

Ai_Ptngan = cv2.VideoWriter_fourcc(*'mp4v')         #بعد كده احفظ الفيديو اللي هتشتغل عليه بصيغه ام بي فور عشان البرنامج يفهمه
out = cv2.VideoWriter("output.mp4", Ai_Ptngan, fps, (width, height))       #output.mp4     # اعمل فيديو جديد بالبيانات اللي قولت عليها الارتفاع و العرض و عدد الفريمات و احفظه في ملف اسمه 

while cap.isOpened():                               #لف علي كل الفريمات و اقرأها طالما الفيديو مفتوح
    ret, frame = cap.read()                     #طب لو حصل اي حاجه في الصوره خلتك ماتعرفش تقراء الفريم اعمل بريك و وقف يعني الفيديو فيه مشكله
    if not ret:
        break

    result = model(frame)       # كل فريم عدي من اللوب بتاعت هوايل تنباء بالبوكس اللي المفروض يبقي علي العربيه او الموتوسكيل
    annotated_result = result[0].plot()     #  هنا بقوله اظهر البوكس اللي تنبأت بيه ده في الفيديو و الصفر عشان يجيب من اول فريم خالص

    out.write(annotated_result)                 # الفيديو الجديد اللي هتعمله خزن فيه البوكس ده لكل اوبجيكت
    cv2.imshow("YOLO", annotated_result)            # بقوله اعرضها في الفيديو البوكسيز

    if cv2.waitKey(1) == ord("q"):              #بقوله لو ضغط علي حرف ال  كيو وقف الفيديو فورا
        break                       # اعمل للفيديو بريك اول لما اضغط كيو

cap.release()                       #
out.release()
cv2.destroyAllWindows()
```
